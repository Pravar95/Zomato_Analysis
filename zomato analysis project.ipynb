```python
import pandas as pd
```

```python
orders = pd.read_csv("Orders.csv")
customers = pd.read_csv("Customers.csv")
restaurants = pd.read_csv("Restaurants.csv")
```

```python
orders["Order_Date"] = pd.to_datetime(orders["Order_Date"], errors='coerce')
```

```python
orders["Month"] = orders["Order_Date"].dt.strftime("%B")
orders
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up the plot
plt.figure(figsize=(6, 3))
sns.countplot(x="Month", data=orders)
plt.xticks(rotation=90)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up the plot
plt.figure(figsize= (6,3))
sns.countplot(x = "Month", data = orders, hue = "Month", palette = 'viridis')
plt.xticks(rotation = 45)
plt.show()
```

```python
orders["Month"] = orders["Order_Date"].dt.strftime("%H")
```

```python
orders.info()
```

```python
orders["Hour"] = orders["Order_Date"].dt.strftime("%H")
orders["Hour"] = orders["Hour"].astype("int")
orders = orders.sort_values(by = "Hour")
sns.lineplot(x = "Hour", y = "Total_Amount", data = orders, estimator = "sum")
plt.show()
```

```python
df = pd.merge(left = orders, right = customers, on = "Customer_ID" , how = "inner")
df
```

```python
sns.countplot(x = "Customer_Location", data = df)
plt.show()
```

```python
plt.figure(figsize = (10,4))
sns.countplot(x = "Customer_Location", data = df)
plt.show()
```

```python
plt.figure(figsize = (10,4))
sns.countplot(x = "Customer_Location", data = df , hue = "Dish_Name")
plt.show()
```

```python
plt.figure(figsize = (10,4))
sns.countplot(x = "Customer_Location", data = df , hue = "Dish_Name")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', title="Dish Name")
plt.tight_layout()  # Adjust layout to prevent clipping
plt.show()
```
